```python
import pymupdf  # import PyMuPDF

if not hasattr(pymupdf.Page, "find_tables"):
    raise RuntimeError("This PyMuPDF version does not support the table feature")
```

```python
doc = pymupdf.open("pdfs/sample.pdf")  # open a document.
```

```python
from page_processor import PageType

for page in doc:
    # get all text with fontsize larger than 20 pt
    blocks = page.get_text("dict")["blocks"]
    for b in blocks:
        if "lines" in b:  # this block contains text
            for l in b["lines"]:  # iterate through the text lines
                for s in l["spans"]:  # iterate through the text spans
                    if s["size"] > 20:  # if fontsize > 20 pt, a title
                        for pt in PageType:
                            if pt.value.lower() in s["text"].lower():
                                print(f"Page {page.number + 1} is of type {pt.name}")
```

```python
# print document toc as a tree
toc = doc.get_toc(simple=False)
for t in toc:
    print("  " * (t[0] - 1) + f"{t[1]} (page {t[2]})")
```
